```python
#Retirando duplicidades

emprestimos = emprestimos.drop_duplicates()
```

```python
#Incrementando dados
exemplares = pd.read_parquet('https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet')
exemplares
```

```python
#Vamos usar Merge para incrementar os DataFrames

completo = emprestimos.merge(exemplares)
```

```python
completo.head()
```

```python
# Localização compreende um número que será adicionado na coluna CDU, que define os temas
#000 a 099: Generalidades. Ciência e conhecimento.
#100 a 199: Filosofia e psicologia.
#200 a 299: Religião.
#300 a 399: Ciências sociais.
#400 a 499: Classe vaga. Provisoriamente não ocupada.
#500 a 599: Matemática e ciências naturais.
#600 a 699: Ciências aplicadas.
#700 a 799: Belas artes.
#800 a 899: Linguagem. Língua. Linguística.
#900 a 999: Geografia. Biografia. História."

CDU_lista = []
for CDU in completo['localizacao']:
  if(CDU < 100):
    CDU_lista.append('Generalidades')
  elif(CDU < 200):
    CDU_lista.append('Filosofia e psicologia')
  elif(CDU < 300):
    CDU_lista.append('Religião')
  elif(CDU < 400):
    CDU_lista.append('Ciências sociais')
  elif(CDU < 500):
    CDU_lista.append('Classe vaga')
  elif(CDU < 600):
    CDU_lista.append('Matemática e ciências naturais')
  elif(CDU < 700):
    CDU_lista.append('Ciências aplicadas')
  elif(CDU < 800):
    CDU_lista.append('Belas artes')
  elif(CDU < 900):
    CDU_lista.append('Linguagem')
  else:
    CDU_lista.append('Geografia. Biografia. História.')

completo['CDU_geral'] = CDU_lista
```

```python
data_emprestimos= pd.DataFrame(completo['data_emprestimo'].value_counts()).reset_index()
data_emprestimos.columns = ['data','quantidade']
data_emprestimos['data']= pd.to_datetime(data_emprestimos['data'])
data_emprestimos
```

```python
# Aqui preciso filtrar anualmente (12 meses) a quantidade de exemplares para podermos ter um melhor mapa

y_emprestimo = data_emprestimos.groupby(by=data_emprestimos.data.dt.year).sum() #Extraímos o ano de todos os empréstimos e efetuamos a soma
y_emprestimo.index.name = 'ano'
y_emprestimo
```

```python
#Plotando


sns.set_theme(style='darkgrid', 
              palette='deep', 
              font_scale=1.3, 
              rc={"figure.figsize":(15,8)})

grafico1 = sns.lineplot(data=y_emprestimo, x='ano', y='quantidade')
grafico1.set_title('Quantidade de exemplares emprestados por ano', size=18, loc='left', weight='bold')
```

```python
#Filtrando empréstimos por mês

m_emprestimos = data_emprestimos.groupby(by=data_emprestimos.data.dt.month).sum()
m_emprestimos.index.name = 'mes'
m_emprestimos
```

```python
grafico2 = sns.lineplot(data=m_emprestimos, x='mes', y='quantidade')
grafico2.set_title('Quantidade de exemplares emprestados por mês de 2010 a 2020', size=18,loc='left',weight='bold')
grafico2=grafico2
```

```python
#Como não apareceram todos os meses, vou fazer um dicionário e converter estes tipos

meses_do_ano={1: 'Janeiro',
              2: 'Fevereiro',
              3: 'Março',
              4: 'Abril',
              5: 'Maio',
              6: 'Junho',
              7: 'Julho',
              8: 'Agosto',
              9: 'Setembro',
              10: 'Outubro',
              11: 'Novembro',
              12: 'Dezembro'
              }

m_emprestimos.index = m_emprestimos.index.map(meses_do_ano)
m_emprestimos
```

```python
# Agora o gráfico vai funcionar melhor

grafico2 = sns.lineplot(data=m_emprestimos, x='mes', y='quantidade')
grafico2.set_title('Quantidade de exemplares emprestados por mês de 2010 a 2020', size=18,loc='left',weight='bold')
grafico2=grafico2
```

```python
horario_pico = data_emprestimos.groupby(by=data_emprestimos.data.dt.hour).sum()
horario_pico.index.name='horas'
horario_pico = horario_pico.reset_index()
horario_pico  
```

```python
# Gerando um gráfico de barras que exibe os horários de pico nos empréstimos

grafico3 = sns.barplot(data= horario_pico, y='quantidade', x='horas', palette='Blues',
                      hue='quantidade', dodge=False)
grafico3.set(xlabel='Horário',
             ylabel=None)
grafico3.set_title('Quantidade de exemplares emprestados por horário', size=18, loc='left', weight='bold')
grafico3=grafico3
```
